```python
def get_financials_of_state_from_excel(excel,state):
    df=excel_file.parse(state,usecols='B:M')
    df=df.T
    df.reset_index(inplace=True,drop=True)
    df=df.drop(labels=[0],axis=1)
    df.columns=df.iloc[0]
    df=df.drop(labels=[0],axis=0)
    df=df.dropna(axis=1,how='all')
    names=df.columns.to_list()
    names[0]='Ano'
    names[names.index('Despesa Total¹')] = 'Despesa Total'
    df.columns=names
    df['estado']=state
    return df

df_RJ=get_financials_of_state_from_excel(excel_file,'RJ')
df_RJ.T
```

```python
def get_financial_per_state_from_excel(excel):
    states=[sheet_name for sheet_name in excel.sheet_names if len(sheet_name)==2 ]
    df=None
    for state in states :
        df_aux=get_financials_of_state_from_excel(excel,state)
        if not (df is None):
            df=df.append(df_aux, ignore_index=True)
        else:
            df=df_aux
    return df


df=get_financial_per_state_from_excel(excel_file)
df['lucro']=df['Receita Bruta']-df['Despesa Total']
df['Previdencia/ReceitaCorrente']=df['Resultado Previdenciário (R$ milhões, a preços correntes)']/df['Receita Corrente']
df['Previdencia/Tributária']=df['Resultado Previdenciário (R$ milhões, a preços correntes)']/df['Tributária']
df['Inativos/Ativos']=df['Inativos']/df['Ativos']
df.T
```

```python
def plot_lucro(df,estado):
    filter_estado=df['estado']==estado
    df_estado=df.loc[filter_estado]
    df_estado.plot(kind='line',x='Ano',y=['Receita Bruta','Despesa Total'],figsize=(16,8))
    df_estado.plot(kind='line',x='Ano',y=['lucro'],figsize=(16,8))
plot_lucro(df,'SP')
```

```python
def plot_lucro_bokeh(df,estado):
    filter_estado=df['estado']==estado
    df_estado=df.loc[filter_estado]
    plot = figure(x_axis_label='Ano', height=400, width =800)
    plot.line(x=df_estado['Ano'],y=df_estado['lucro'])
    show(plot)
plot_lucro_bokeh(df,'RJ')
```

```python
filter_ano=df['Ano']==2017
```

```python
any(filter_ano)
```

```python
df_2017=df.loc[filter_ano]
df_2017_organizado_por_RP=df_2017.sort_values(by='Resultado Previdenciário (R$ milhões, a preços correntes)')
plot = figure(x_axis_label='Estado', 
              x_range=df_2017_organizado_por_RP["estado"], 
              height=400, width =800, 
              y_axis_label='Resultado Previdenciario',
              title='Resultado Previdenciario')
plot.vbar(x=df_2017_organizado_por_RP['estado'],top=df_2017_organizado_por_RP['Resultado Previdenciário (R$ milhões, a preços correntes)'], width=0.5,color = "blue")
show(plot)
```

```python
def plot_analises_previdenciario(df, ano):
    filter_ano=df['Ano']==ano
    df_ano=df.loc[filter_ano]
    df_ano_sorted=df_ano.sort_values(by='Resultado Previdenciário (R$ milhões, a preços correntes)')
    plot = figure(x_axis_label='Estado', 
                  x_range=df_ano_sorted["estado"], 
                  height=400, width =800, 
                  y_axis_label='Resultado Previdenciario',
                  title='Resultado Previdenciario')
    plot.vbar(x=df_ano_sorted['estado'],
              top=df_ano_sorted['Resultado Previdenciário (R$ milhões, a preços correntes)'], 
              width=0.5,color = "blue")
    show(plot)
plot_analises_previdenciario(df,2017)
```

```python
def plot_analises_previdenciario_por_medida(df, ano,medida):
    filter_ano=df['Ano']==ano
    df_ano=df.loc[filter_ano]
    df_ano_sorted=df_ano.sort_values(by=medida)
    plot = figure(x_axis_label='Estado', 
                  x_range=df_ano_sorted["estado"], 
                  height=400, width =800, 
                  y_axis_label=medida,
                  title='Resultado Previdenciario: '+ medida)
    plot.vbar(x=df_ano_sorted['estado'],
              top=df_ano_sorted[medida], 
              width=0.5,color = "blue")
    show(plot)
plot_analises_previdenciario_por_medida(df,2017,'Previdencia/ReceitaCorrente') 
```

```python
plot_analises_previdenciario_por_medida(df,2017,'Previdencia/Tributária')
```
